```python
import pandas as pd #For reading dataset files
import networkx as nx #For network creation/analysis
import matplotlib.pyplot as plt #For plotting graphs
import json

original_data = pd.read_csv('artists.csv')
artist_id = []
artist_from = []
artist_to = []
    
f = open('dict_artists.json')

data = json.load(f)
  
print(len(data))
for key, value in data.items():
    
    artist_f = original_data['name'][original_data['id'] == key].values[0]
    
    artist_id.append(artist_f)
    if len(value) > 0:    
        for elem in value:
            artist_from.append(artist_f)
            artist_to.append(original_data['name'][original_data['id'] == elem].values[0])

    
# Closing file
f.close()
```

```python
data = {'artist_from':artist_from,
        'artist_to':artist_to}

artist_from = artist_from.pop()
print(len(artist_from))
print(len(artist_to))

data_nodes = {'id': artist_id}

df_node = pd.DataFrame(data_nodes)
df = pd.DataFrame(data)
df_node
```
